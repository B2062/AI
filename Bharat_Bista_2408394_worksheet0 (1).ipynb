Task 1 - Sum of Nested Lists:


```python
def sum_nested_list(nested_list):
  total = 0
  for element in nested_list:
    if isinstance(element, list):
      total += sum_nested_list(element) # Recursively sum the nested list
    else:
      total += element # Add the number to the total
  return total

#• Test the function with a sample nested list, such as
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)

# Print the result
print("The sum of all numbers in the nested list is:", result)
```

Task 2 - Generate All Permutations of a String:

```python
def generate_permutations(s):
    if len(s) == 1:  # Base case: A single character has only one permutation
        return [s]

    permutations = set()  # Use a set to ensure unique permutations
    for i, char in enumerate(s):
        # Exclude the current character and find permutations of the remaining substring
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            permutations.add(char + perm)

    return list(permutations)  # Convert the set to a list before returning


print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
def calculate_directory_size(directory):
    """
    Recursively calculate the total size of a directory.

    Args:
        directory (dict): A dictionary representing the directory structure.
            Keys are file or directory names, and values are file sizes (int) or nested directories (dict).

    Returns:
        int: Total size of the directory in KB.
    """
    total_size = 0

    for name, value in directory.items():
        if isinstance(value, dict):  # If the value is a nested directory
            total_size += calculate_directory_size(value)
        else:  # If the value is a file size
            total_size += value

    return total_size

# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

# Test the function
total_size = calculate_directory_size(directory_structure)
print("Total size of the directory:", total_size, "KB")
```

```python
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up a given amount using dynamic
programming.
This function solves the coin change problem by determining the fewest number of
coins from a given set of coin denominations that sum up to a target amount. The
solution uses dynamic programming(tabulation) to iteratively build up the minimum
number of coins required for each amount.

    Args:
    coins (list of int): A list of coin denominations available for making change. Each
coin denomination is a positive integer.
amount (int): The target amount for which we need to find the minimum number of coins
. It must be a non-negative integer.

    Returns:
    int: The minimum number of coins required to make the given amount.
If it is not possible to make the amount with the given coins, returns -1.
    """
    # Initialize a DP array where dp[i] represents the minimum coins required to make amount i
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins are needed to make amount 0

    # Iterate through each coin and build the solution for each amount
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)  # Update DP table with the minimum coins

    # If the amount is still 'inf', it means it's not possible to form that amount
    return dp[amount] if dp[amount] != float('inf') else -1

# Test the function
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print("Minimum coins needed to make amount", amount, ":", result)
```

```python
def knapsack(weights, values, capacity):
    """
    Solves the 0/1 Knapsack problem using dynamic programming.

    Args:
        weights (list of int): Weights of the items.
        values (list of int): Values of the items.
        capacity (int): Maximum weight capacity of the knapsack.

    Returns:
        int: The maximum value that can be achieved within the weight capacity.
    """
    n = len(weights)
    # Create a DP table where dp[i][w] represents the maximum value for the first i items with weight limit w
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    # Build the table in a bottom-up manner
    for i in range(1, n + 1):  # Loop through items
        for w in range(1, capacity + 1):  # Loop through capacities
            if weights[i - 1] <= w:
                # Option 1: Include the current item
                include_item = values[i - 1] + dp[i - 1][w - weights[i - 1]]
                # Option 2: Exclude the current item
                exclude_item = dp[i - 1][w]
                # Take the maximum of the two options
                dp[i][w] = max(include_item, exclude_item)
            else:
                # If the item's weight exceeds the current capacity, exclude it
                dp[i][w] = dp[i - 1][w]

    # The result is in dp[n][capacity]
    return dp[n][capacity]

# Test the function
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
result = knapsack(weights, values, capacity)
print("Maximum value within capacity", capacity, ":", result)
```
